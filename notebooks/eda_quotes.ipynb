## Questions Before Analysis

1. How many quotes are in the dataset?
2. How many unique authors are present?
3. Are there missing values?
4. Which authors appear most frequently?
5. What is the typical length of a quote?


```python
import pandas as pd

df = pd.read_csv("../data/quotes_data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
df["Author"].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt

top_authors = df["Author"].value_counts().head(10)

plt.figure()
top_authors.plot(kind="bar")
plt.title("Top 10 Authors by Number of Quotes")
plt.xlabel("Author")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df["Quote_Length"] = df["Quote"].apply(len)
df["Quote_Length"].describe()
```

```python
plt.figure()
plt.hist(df["Quote_Length"], bins=20)
plt.title("Distribution of Quote Lengths")
plt.xlabel("Quote Length")
plt.ylabel("Frequency")
plt.show()
```

## Conclusion

- The dataset is clean with no missing values.
- Some authors contribute significantly more quotes than others.
- Most quotes have moderate length, with a few long outliers.
- The dataset is suitable for further analysis such as visualization or sentiment analysis.
